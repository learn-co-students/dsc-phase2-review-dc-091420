## Question 1
A Hollywood executive wants to know how much a R-Rated movie released after 2000 will earn. The data above is a sample of some of the movies with that ranting during that time frame, as well as other movies. How would you go about answering her question? Talk through it theoretically and then do it in code.

Formula for confidence interval: mean +- std_err*sd_multiplier, (sd_multiplier determines the percentage of the confidence interval)

for 95%, sd_multiplier is 1.96

```python
#your answer here

subset = df[
    (df['content_rating']=='R') &
    (df['title_year'] >= 2000)

]

subset_mean = subset.gross.mean()
subset_std = subset.gross.std()
n = len(subset)

std_err = subset_std / (n**.5)

low_conf_int = subset_mean-std_err*1.96
high_conf_int = subset_mean+std_err*1.96

print(f"mean: {subset_mean}")
print(f"conf int: {low_conf_int}, {high_conf_int}")
```

## Question 2b

Calculate the answer for the second question
- Do foreign films perform differently at the box office than non-foreign films?

```python
#your answer here
from scipy.stats import ttest_ind
from scipy import stats

'''
H_null: the means of foreign and domestic films do not significantly differ
H_alt: the means of foreign and domestic films do significantly differ
alpha level= .05

one-sided or two-sided: two-sided
'''


foreign = df[df['language'] != 'English']
domestic = df[df['language'] == 'English']


ddf = (len(foreign)+len(domestic))-2

crit_below = stats.t.ppf(.025, ddf)
crit_above = stats.t.ppf(.975, ddf)

for_gross = foreign.gross
dom_gross = domestic.gross

for_gross.dropna(inplace=True)
dom_gross.dropna(inplace=True)

print(crit_below, crit_above)
ttest_ind(for_gross, dom_gross, equal_var=False)
```

## Question 4a

Create the following variables:
- `years_old`: The number of years since the film was released
- Dummy categories for each of the following ratings:
  - `G`
  - `PG`
  - `R`

Once you have those variables, create a summary output for the following OLS model:

`gross = cast_total_facebook_likes + budget + years_old + G + PG + R`

```python
#your answer here

model_data['years_old'] = 2020 - model_data['title_year']

ratings = (
    pd.get_dummies(
        model_data, columns=['content_rating']
    )
    .drop(columns='content_rating_PG-13')
)

model_frame = pd.concat([model_data, ratings])

model = ols(formula="gross~cast_total_facebook_likes + budget + years_old + content_rating_G + content_rating_PG + content_rating_R", data=model_frame).fit()
model.summary()

# model_frame.sample(5)
```
